```python
def prepare_energy(energy_data, nc_energy, ref_energy):
    X, Y = zip(*energy_data)
    mask = np.where(np.array(Y))[0]
    Y = np.append(nc_energy, np.array(Y)[mask]) - ref_energy
    X = np.append(0, np.array(X)[mask])
    return X, Y

def prepare_score(score_data):
    X, Y = zip(*score_data)
    X = np.append(0, X)
    Y = np.append(0, Y)
    return X, Y
    
def plot_data(data, energy_axis, legacy=None):
    nc_energy  = data['noncon_energy']
    hf_energy = data['hf_energy']
    mp2_energy = data['mp2_energy']
    ccsd_energy = data['ccsd_energy']
    ref_energy = data['fci_energy']
    
    X_Hc, Y_Hc = prepare_energy(data['data']['cs_vqe_nrg_Hc'], nc_energy, ref_energy)
    X_CC, Y_CC = prepare_energy(data['data']['cs_vqe_nrg_CC'], nc_energy, ref_energy)
    X_MP, Y_MP = prepare_energy(data['data']['cs_vqe_nrg_MP'], nc_energy, ref_energy)
    X_HL, Y_HL = prepare_energy(data['data']['cs_vqe_nrg_HL'], nc_energy, ref_energy)
    max_X = X_Hc[-1]
    
    energy_axis.plot(X_Hc, Y_Hc, color='cyan', marker='o', label='$H_c$-motivated')
    energy_axis.plot(X_HL, Y_HL, color='blue', marker='o', label='HOMO-LUMO-biased')
    energy_axis.plot(X_MP, Y_MP, color='pink', marker='o', label='MP2-motivated')
    energy_axis.plot(X_CC, Y_CC, color='purple', marker='o', label='CC-motivated')
    if legacy is not None:
        Y_legacy = np.array(legacy[:-1])# - ref_energy
        X_legacy = range(len(Y_legacy))
        energy_axis.plot(X_legacy, Y_legacy, color='black', marker='.', label='Original', ls=':')
    
    energy_axis.hlines(hf_energy-ref_energy, 0, max_X, ls=':', color='red', label='HF energy')
    energy_axis.hlines(mp2_energy-ref_energy, 0, max_X, ls=':', color='orange', label='MP2 energy')
    energy_axis.hlines(ccsd_energy-ref_energy, 0, max_X, ls=':', color='green', label='CCSD energy')
    energy_axis.hlines(0.0016, 0, max_X, ls=':', color='black', label='Chemical accuracy')
    energy_axis.set_yscale('log')
    energy_axis.set_title(speciesname)
    energy_axis.set_xticks(X_Hc)
    
    score_axis = energy_axis.twinx()
    X_Hc_score, Y_Hc_score = prepare_score(data['data']['stab_score_Hc'])
    X_CC_score, Y_CC_score = prepare_score(data['data']['stab_score_CC'])
    X_MP_score, Y_MP_score = prepare_score(data['data']['stab_score_MP'])
    X_HL_score, Y_HL_score = prepare_score(data['data']['stab_score_HL'])
    
    score_axis.plot(X_Hc_score, Y_Hc_score, color='cyan', label='$H_c$-motivated score', marker='.', alpha=0.2, ls='--')
    score_axis.plot(X_Hc_score, Y_HL_score, color='blue', label='HOMO-LUMO-biased score', marker='.', alpha=0.2, ls='--')
    score_axis.plot(X_MP_score, Y_MP_score, color='pink', label='MP2-motivated score', marker='.', alpha=0.2, ls='--')
    score_axis.plot(X_CC_score, Y_CC_score, color='purple', label='CC-motivated score', marker='.', alpha=0.2, ls='--')
```
